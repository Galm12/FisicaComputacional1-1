En este programa se calcula las áreas de un **círculo** y de una **elipse**, para el caso del círculo:
$$A=\pi*r^2$$
La variable **r** representa el radio de la circunferencia. En el caso de la elipse:
$$A=\pi*a*b$$
Las variables **a** y **b** son los semiejes mayor y menor correspondientemente de la elipse. 

```
  import numpy as np

  def area1(x):
    return np.pi * x ** 2
  def area2(x2,y):
    return np.pi * x2 * y

  print("Este programa calcula las áreas para un círculo y")
  print("una elipse")
  print("Por lo tanto presiona c para calcular la del círculo y e para la elipse:",end="")

respuesta=input()

if respuesta == "c" :
   print ( "Se requiere el valor del radio: ",  end="" )
   r = float(input()) 
   a1 = area1(r)
   print ( "El área del círculo es {:.2f}" .format(a1))

elif respuesta == "e":
    print ( "Se requiere el valor del eje mayor y del eje menor" )
    a = float(input()) 
    b = float(input())
    a2 = area2(a,b)
    print ( "El área de la elipse es {:.2f}" .format(a2))
```

Este programa calcula los volumenes de una **esfera** y de un **cilindro circular**:

El volumen de la esfera se calcula de la siguiente manera:
$$V=\frac{4}{3}*\pi*r^3$$
Donde **r** representa el radio de la esfera, para el caso del volumen del cilindro circular:
$$V=\pi*r^2*h$$
Donde **r** representa el radio de la circunferencia base y **h** la altura del cilindro.

```
  import numpy as np

  def volumene(x):
    return (4/3) * np.pi * x ** 3
  def volumencc(x,y):
    return y * np.pi * x ** 2

  print("Este programa calcula el volumen de una esfera y")
  print("de un cilindro circular")
  print("Por lo tanto presiona e para calcular la de esfera y c para el cilindro circular:",end="")

respuesta=input()

if respuesta == "e" :
   print ("Se requiere el valor del radio: ",  end="" )
   r = float(input()) 
   v1 = volumene(r)
   print ( "El volumen de la esfera es {:.2f}" .format(v1))

elif respuesta == "c":
    print ( "Se requiere el valor del radio y de la altura" )
    a = float(input()) 
    b = float(input())
    v2 = volumencc(a,b)
    print ( "El volumen del cilindro es {:.2f}" .format(v2))
```

```
import numpy as np

def discriminante(x,y,z):
    return ((y ** 2) - (4 * x * z))

def solucion1(x,y,z):
    return ((-1 * y) + np.sqrt((y ** 2) - (4 * x * z))) / (2 * x)

def solucion2(x,y,z):
    return ((-1 * y) - np.sqrt((y ** 2) - (4 * x * z))) / (2 * x)

def solucion3(x,y):
    return ((-1 * y) / (2 * x))

print("Este programa calcula las raíces de una ecuación cuadrática")
print("Ingrese el valor del término cuadrático")
print("Ingrese el valor del término lineal")
print("Ingrese el valor del término independiente")
A = float(input())
B = float(input())
C = float(input())

D = float(discriminante(A,B,C))

if D > 0:
   r1 = solucion1(A,B,C)
   r2 = solucion2(A,B,C)
   print ( "Las raíces son:")
   print ( "Primera raíz", r1)
   print ( "Segunda raíz", r2)
if D < 0:
   print ( "La solución es compleja" )
if D == 0:
   r3 = solucion3(A,B)
   print ( "La solución es:", r3 )
```

```
import numpy as np

def raíz_cuadrada_mediante_el_método_Babilonio(n):

    if n < 0:
       return print ( "No se puede con números negativos" )
    
    if n == 0:
       return print ( "La raíz es igual a cero" )

    else:
         x = 2.0
         cociente = n / x
         promedio = (x + cociente) / 2.0

         while abs(cociente - promedio) > 0.01:
              
                x = promedio
                cociente = n / x
                promedio = (x + cociente) / 2.0

         return promedio

print("Ingrese el número que se quiere sacar su raíz cuadrada")

N = float(input())

print ( "Aproximación con el método Babilonio:", raíz_cuadrada_mediante_el_método_Babilonio(N) )
print ( "Valor real de la raíz", np.sqrt(N) )
```

```
# Cargar la subbiblioteca pytlot de Matplotlib 

import matplotlib.pyplot as plt
import numpy as np

# Gráfica de funciones
# Definición de variables y funciones

x = np.linspace(-np.pi,np.pi,100)  # arrego de 100 puntos entre -pi y pi

y = np.log(x+1)      # y será también un arreglo de 100 puntos automáticamente 

y1 = x

y2 = x**2/(2)

y3 = x**3/(3)

y4 = x**4/(4)

y5 = x**5/(5)

y6 = x**6/(6)

y7 = x**7/(7)

y8 = x**8/(8)

y9 = x**9/(9)

y10 = x**10/(10)

y11 = x**11/(11)

y12 = x**12/(12)

y13 = x**13/(13)

y14 = x**14/(14)

y15 = x**15/(15)

y16 = x**16/(16)

sT1 = y1 - y2 + y3 - y4 # Serie de Taylor de ln(x+1) a 4o. Orden.
sT2 = y1 - y2 + y3 - y4 + y5 - y6 + y7 # Serie de Taylor de ln(x+1) a 7o. Orden.
sT3 = y1 - y2 + y3 - y4 + y5 - y6 + y7 - y8 + y9 - y10 + y11 # Serie de Taylor de ln(x+1) a 11o. Orden.
sT4 = y1 - y2 + y3 - y4 + y5 - y6 + y7 - y8 + y9 - y10 + y11 - y12 + y13 - y14 + y15 - y16 # Serie de Taylor de ln(x+1) a 16o. Orden.


# Creación de la gráficas

plt.subplot(111)

plt.xlabel("X label")

plt.ylabel("Y label")

plt.title("Title")


# Graficar varias funciones

plt.plot(x,y, label="log(x+1)")

plt.plot(x,sT1, label='Serie Taylor Orden 4')

plt.plot(x,sT2, label='Serie Taylor Orden 7')

plt.plot(x,sT3, label='Serie Taylor Orden 11')

plt.plot(x,sT4, label='Serie Taylor Orden 16')

# Agregar una malla

plt.grid(True)

# Fijar los límites en el eje X.

plt.xlim(-1.5, 1.5)

# Fijar los límites en el eje Y.

plt.ylim(-4, 2)

# Ubicación de las etiquetas de las funciones

#plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.legend(loc="upper left")

plt.show()

#plt.savefig('plot.png')  # Si se desea guardar en un archivo la imagen de la gráfica

 
```

```
Gráficas de funciones en 2D

 

# Cargar la subbiblioteca pytlot de Matplotlib 

import matplotlib.pyplot as plt


# Gráfica de funciones
# Definición de variables y funciones

x = np.linspace(-np.pi,np.pi,100)  # arrego de 100 puntos entre -pi y pi

y = np.sin(x)      # y será también un arreglo de 100 puntos automáticamente 

y1 = x

y3 = x**3/(3*2)

y5 = x**5/(5*4*3*2)

sT = y1 - y3 + y5 # Serie de Taylor de sin(x) a 5o. Orden.

 

# Creación de la gráficas

plt.subplot(111)

plt.xlabel("X label")

plt.ylabel("Y label")

plt.title("Title")


# Graficar varias funciones

plt.plot(x,y, label="sin(x)")

plt.plot(x,y1, label='x')

plt.plot(x,sT, label='Serie Taylor Orden 5')

# Agregar una malla

plt.grid(True)

# Fijar los límites en el eje Y.

plt.ylim(-2, 2)

# Ubicación de las etiquetas de las funciones

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()

#plt.savefig('plot.png')  # Si se desea guardar en un archivo la imagen de la gráfica
```
